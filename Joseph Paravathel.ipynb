```python
sns.scatterplot(x='SL', y='SW',hue='Classification', data=data )
plt.show()
```

```python
### 1.Species Setosa has smaller sepal lengths but larger sepal widths.
### 2.Versicolor Species lies in the middle of the other two species in terms of sepal length and width
### 3.Species Virginica has larger sepal lengths but smaller sepal widths.
```

```python
sns.stripplot(data['Classification'],data['PL'])
plt.show()
```

```python
### 4.The petal length of iris setosa is ranging from 1 cm to 2 cm and it is distinct from iris - versicolor and iris virginica
```

```python
sns.stripplot(data['Classification'],data['PW'])
plt.show()
```

```python
### 5.The petal width of iris setosa is ranging 0.2 cm to 0.6 cm ,also it is distinct from rest of the iris species.
```

```python
### 6.From both these strip plot we can understand that the petal size of iris setosa is very small compared to others.
### 7.And it is a distinctive feature 
```

```python
sns.heatmap(data.corr(),annot=True)
plt.show()
```

```python
### 8.Petal width and petal length have high correlations. 
```

```python
plt.figure(figsize=(6,5))

plt.subplot(2,2,1)
plt.hist(data['SL'])
plt.title('Histogram1')

plt.subplot(2,2,2)
plt.hist(data['SW'])
plt.title('Histogram2')

plt.subplot(2,2,3)
plt.hist(data['PL'])
plt.title('Histogram3')

plt.subplot(2,2,4)
plt.hist(data['PW'])
plt.title('Histogram4')


plt.tight_layout()
plt.show()
```

```python
### 9.The highest frequency of the sepal length is between 30 and 35 which is between 5.5 and 6
### 10.The highest frequency of the sepal Width is around 70 which is between 3.0 and 3.5
### 11.The highest frequency of the petal length is around 50 which is between 1 and 2
### 12.The highest frequency of the petal width is between 40 and 50 which is between 0.0 and 0.5
```

```python
### 13.In Histogram 3 and Histogram 4 we can see two classes .
### 14.That is petal length and petal width of iris setosa is distinct from others.
```

```python
plt.figure(figsize=(6,5))

plt.subplot(2,2,1)
sns.boxplot(x="Classification", y='SL', data=data)
plt.xticks(rotation=45)

plt.subplot(2,2,2)
sns.boxplot(x="Classification", y='SW', data=data)
plt.xticks(rotation=45)

plt.subplot(2,2,3)
sns.boxplot(x="Classification", y='PL', data=data)
plt.xticks(rotation=45)

plt.subplot(2,2,4)
sns.boxplot(x="Classification", y='PW', data=data)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
